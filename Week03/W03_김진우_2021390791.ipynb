```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head()
```

```python
plt.figure(figsize=(6, 5))
sns.countplot(x='Attrition', data=df)
plt.title('Attrition Distribution')
plt.ylabel('Number of Employees')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.title('Attrition by OverTime')
plt.ylabel('Number of Employees')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(y='JobRole', hue='Attrition', data=df)
plt.title('Attrition by Job Role')
plt.xlabel('Number of Employees')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='MaritalStatus', hue='Attrition', data=df)
plt.title('Attrition by Marital Status')
plt.ylabel('Number of Employees')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='MaritalStatus', hue='Attrition', data=df)
plt.title('Attrition by Marital Status')
plt.ylabel('Number of Employees')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['MonthlyIncome'], bins=30, kde=True)
plt.title('Distribution of Monthly Income')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
order = df.groupby('JobRole')['MonthlyIncome'].median().sort_values().index
sns.boxplot(y='JobRole', x='MonthlyIncome', data=df, order=order)
plt.title('Monthly Income by Job Role')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.countplot(x='JobSatisfaction', hue='Attrition', data=df, ax=axes[0])
axes[0].set_title('Attrition by Job Satisfaction')

sns.countplot(x='EnvironmentSatisfaction', hue='Attrition', data=df, ax=axes[1])
axes[1].set_title('Attrition by Environment Satisfaction')

plt.tight_layout()
plt.show()
```
